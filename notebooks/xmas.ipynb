I want to make a Christmas tree on my wall of fairy lights. The triangle is 103cm across the bottom and 120cm down each side. I have 19 metres of lights. Can anyone advise how far apart I need to space the nails to evenly cover the tree zig zagging the lights from the centre top then left to right repeat repeat repeat.

```python
import math
def c(i,x):
    """
    i - index of zigzag's base
    x - number of parts to divide the side into
    returns length of the zigzag
    """
    a = 120/x
    b = i*103/x
    c_squared = a**2 + b**2 - 2*a*b*math.cos(math.radians(64.5))
    return math.sqrt(c_squared)
```

```python
def tot_length(parts):
    x = parts
    return 120/x + 103/x + sum([c(i,x)+(i*103/x) for i in range(2,x+1)])
```

```python
tot_length(17)
```

What an interesting problem!

I approached it by setting x to be the number of parts to divide a side into. Thus the nails would be 120/x cm apart. This means that the length of the first two lines (see picture) is 120/x + 103/x. Now to the interesting part .... the zigzags represented by c.

Using trig, we can find that the top angle is 51°, making the side angles 64.5°. So for the triangle going down, we have the zigzag c, with base 2*103/x and short side 120/x. The next zigzag will have base 3*103/x and short side 120/x, and so on... So the total length of the lights will be
(120/x + 103/x) + (c_2 + 2*103/x) + (c_3 + 3*103/x) + ...

Finding each c isn't hard using the law of cosines, i.e.
c² = A²+B²-2ABcos(64.5°), thus
c = sqrt(A²+B²-2ABcos(64.5°))

However, remember that A and B are 120/x and n*103/x and now buried in multiple square roots, so solving for x algebraically is hard if not impossible.

So I cheated. I wrote a computer program to calculate the total wire length based on the number of parts. I fed it different numbers until it got just below 1900 cm.

Answer turned out to be 17 parts, which takes up 18.15 meters of lights. Thus, the distance between the nails should be 120/17 = 7.06 cm apart. Call it 7cm, and you should have just under a meter of spare lights.
