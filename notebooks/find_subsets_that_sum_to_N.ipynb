```python

nums = [
    9.73, 42.87, 74.32, 30.37, 
    107.23, 35.38, 129.34, 25.73, 
    170.53, 170.53, 225.21, 104.55, 
    52.01, 69.70, 129.76, 68.60, 
    7.48, 7.48, 32.71, 93.30, 
    30.37, 48.25, 74.32, 107.24]
    
target = 826.82
#target = 1020.20
best = 10000000000
diff = abs(target-best)

# iterate through 2^len(nums).
for i in range(0,int(2**len(nums))):
    
    # turn i into len(nums) bits that determine which nums to sum
    # 0 - pad with zeros
    # 24 - length of string
    # b - binary representation
    format_code = '0' + str(len(nums)) + 'b' 
    flags = format(i, '024b')
    
    # get list of (flag, num) tuples
    zipped = list(zip(flags,nums))
    
    # sum the numbers indicated by the flag
    selected = [v for idx,v in zipped if idx=='1']
    candidate = sum(selected)
    
    # see if were are any closer than previous best
    if abs(candidate-target) < diff:
        best = candidate
        diff = abs(target-best)
        print(f'{best}:{selected}')
    if diff < 0.001: break #do this because floats are approximate
```
